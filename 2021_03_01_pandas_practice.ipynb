```
df.set_index('datetime', inplace=True)
df.head()
```

```
df.rename(columns={'y': 'sales'}, inplace=True)
df.head()
```

```
df.sort_values(['sales', 'temperature'], ascending=False).head() # ascending=Falseで昇順
```

```
df.index = pd.to_datetime(df.index, format='%Y-%m-%d')
df.index
```

```
df.sort_index().head()
```

```
df.resample('M').mean()
```

```
df['month'] = list(pd.Series(df.index).apply(lambda x: x.month))
df['day'] = list(pd.Series(df.index).apply(lambda x: x.day))

labels = ['上旬', '中旬', '下旬']
df['period'] = pd.cut(list(df['day']),  bins=[0,10,20,31], labels=labels, right=True)

df.head()
```

```
df.isnull().sum()
```

```
df.fillna(value={'payday': 0.0}, inplace=True)
df.head()

df.isnull().sum()
```

```
df['precipitation'] = df['precipitation'].astype(str).str.replace('--', 0).astype(float)
df.head()
```
